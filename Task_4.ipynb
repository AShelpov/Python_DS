```python
df = pd.read_csv("creditcard.csv")
```

```python
list_of_columns = list(df.columns)
list_of_columns.remove("Time")
list_of_columns.remove("Amount")
list_of_columns.remove("Class")
```

```python
list_of_df = []
for column in list_of_columns:
    array = np.array([df[column].value_counts(normalize=True).index.to_numpy(), 
                      df[column].value_counts(normalize=True).values])
    array = array.transpose()
    name = f"df_{column}"
    name = pd.DataFrame(array, columns=pd.MultiIndex.from_arrays(np.array([[column, column],["Values", "Freq"]])))
    list_of_df.append(name)
```

```python
df_counts = pd.DataFrame()
for temp_df in list_of_df:
    df_counts = pd.concat([df_counts, temp_df], axis=1)
```

```python
X = df.loc[:, :"Amount"].copy()
```

```python
y = pd.Series(df["Class"].copy())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100, stratify=y)
```

```python
print(f" X_train {X_train.shape} \n X_test {X_test.shape}  \n y_train {y_train.shape} \n y_test {y_test.shape}" )
```

```python
parametrs = [{
    "n_estimators": [10, 15],
    "max_features": np.arange(3, 5),
    "max_depth": np.arange(4,7)
}]
```

```python
clf = GridSearchCV(estimator=RandomForestClassifier(random_state=100),
                   param_grid=parametrs,
                   scoring="roc_auc",
                   cv=3)
```

```python
clf.fit(X_train, y_train)
```

```python
clf.best_params_
```

```python
y_pred_proba = clf.predict_proba(X_test)
```

```python
y_pred_proba = y_pred_proba.transpose()[1]
```

```python
y_pred_proba
```

```python
y_test = y_test.to_numpy()
```

```python
from sklearn.metrics import roc_auc_score
```

```python
AUC = roc_auc_score(y_test, y_pred_proba)
```

```python
AUC
```

AUC на уровне 0.9462664156037156, говорит о том, что полученная модель получилась с хорошей предсказательной силой и корректно определяет True Positive и False Negative
